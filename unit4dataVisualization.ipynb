# Data Visualization for various data scales

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.array(['Q1', 'Q2', 'Q3', 'Q4'])
y = np.array([243.0, 254.7, 144.6, 317.6])
plt.bar(x,y,color='cyan') # main line 
plt.title('BAR CHART Jeevan Koiri')
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x, bins=20, color='green', density=True)
plt.title('Histogram')
plt.show()
```

```python
x = np.random.randn(10000)
plt.hist(x, bins=100, color='green', density=True, alpha=0.3)
plt.title('Histogram')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

fig0 = plt.figure()
axes0 = fig0.add_axes([0,0,1,1]) # left buttom width height
axes0.plot(x,y, color='red', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
axes0.set_xlim([0,5]) # horizontal axis limit 
axes0.set_ylim([-2,2]) # vertical axis limit 
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

fig0 = plt.figure(figsize=(10,1)) # width and height specified with figsize

axes0 = fig0.add_axes([0,0,1,1]) # left buttom width height
axes0.plot(x,y, color='red', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
axes0.set_xlim([0,5]) # horizontal axis limit 
axes0.set_ylim([-2,2]) # vertical axis limit 
plt.show()
```

```python
fig0 = plt.figure(figsize=(5,2), dpi=100) # width and height specified with figsize

axes0 = fig0.add_axes([0,0,1,1]) # left buttom width height
axes0.plot(x,y, color='orange', linestyle='--', label='Sin')
axes0.plot(x,z, color='purple', linestyle='--', label='Cos')

axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.legend(loc=0) # top right index of the name as in label 
axes0.set_title('LINE PLOT')
# axes0.set_xlim([0,5]) # horizontal axis limit 
# axes0.set_ylim([-2,2]) # vertical axis limit 
plt.show()
```

```python
fig0 = plt.figure(figsize=(8,3), dpi=80) # width and height specified with figsize

axes1 = fig0.add_axes([0,0,0.3,1]) # left buttom width height
axes2 = fig0.add_axes([0.5, 0, 0.3, 1])
axes1.plot(x,y, color='orange', linestyle='--', label='Sin')
axes2.plot(x,z, color='purple', linestyle='--', label='Cos')

axes1.set_xlabel('X')
axes1.set_ylabel('Y')
# axes0.legend(loc=0) # top right index of the name as in label 
axes1.set_title('LINE PLOT')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title("COSINE")
# axes0.set_xlim([0,5]) # horizontal axis limit 
# axes0.set_ylim([-2,2]) # vertical axis limit 
plt.show()
```

```python
fig0 = plt.figure(figsize=(8,4), dpi=80) # width and height specified with figsize

axes1 = fig0.add_axes([0,0.5,0.2,0.5]) # left buttom width height
axes2 = fig0.add_axes([0.5, 0, 0.2, 0.5])
axes1.plot(x,y, color='orange', marker='^', label='Sin', markerfacecolor='purple', markeredgecolor='red', alpha=0.5)
axes2.plot(x,z, color='purple', linestyle='--', label='Cos')

axes1.set_xlabel('X')
axes1.set_ylabel('Y')
# axes0.legend(loc=0) # top right index of the name as in label 
axes1.set_title('COSINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title("SINE")
# axes0.set_xlim([0,5]) # horizontal axis limit 
# axes0.set_ylim([-2,2]) # vertical axis limit 
plt.show()
```
