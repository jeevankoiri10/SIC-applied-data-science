```python
# optimal data exploration merging and binding dataframes 
```

```python
col_data_A = ['Name', 'Gender', 'Age']
data_A = [['Harry Potter', 'Male', 23],['David Baker', 'Male', 23],['John Smith', 'Male', 22],['Juan Martinez', 'Male', 36],['Jane Connor', 'Male', 23]]
df_A = pd.DataFrame(columns=col_data_A, data=data_A)
df_A
```

```python
col_data_B = ['Name', 'Position', 'Wage']
data_B = [['John Smith', 'Intern', 25000],['Alex Du Bois', 'Team Lead', 75000],['Joanne Rowling', 'Manager', 90000], ['Jane Connor', 'Manager', 70000]]
df_B = pd.DataFrame(columns=col_data_B, data=data_B)
df_B
```

```python
# inner join
pd.merge(df_A, df_B, on="Name")
```

```python
pd.merge(df_A, df_B, left_on='Name', right_on='Name', how='inner')
```

```python
pd.merge(df_A, df_B, left_on='Name', right_on='Name', how='left')
```

```python
pd.merge(df_A, df_B, left_on='Name', right_on='Name', how='right')
```

```python
pd.merge(df_A, df_B, left_on='Name', right_on='Name', how='outer')
```

```python
# bind vertically by matching the columns name
pd.concat([df_A, df_B], sort=True)
```

```python
# bind horizontally by matching the columns name
pd.concat([df_A, df_B], axis=1, sort=True)
```

# Hierarchical indexing with multiindexing

```python
import numpy as np
import pandas as np
```

```python
my_header = ['a', 'b', 'c']
my_index_out = ['G1']*3 + ['G2']*3
my_index_in = [1,2,3]*2
my_index_zipped = list(zip(my_index_out, my_index_in))
my_index = pd.MultiIndex.from_tuples(my_index_zipped)
df = pd.DataFrame(data = np.random.randn(6,3), index=my_index, columns=my_header)
df
```
